```python
aux = pd.DataFrame({"Model":["FayOK"],
           "Capacity (mAh)":[8000],
           "Cells":[3],
           "Discharge Rate":[40],
           "Volume (mm^3)":[262080],
           "Weight (g)":[595]})
df = df.append(aux)
df
```

```python
aux = pd.DataFrame({"Model":["YOWOO Graphene Battery"],
           "Capacity (mAh)":[5000],
           "Cells":[3],
           "Discharge Rate":[100],
           "Volume (mm^3)":[171120],
           "Weight (g)":[451.5]})
df = df.append(aux)
aux = pd.DataFrame({"Model":["ZIPPY Flightmax"],
           "Capacity (mAh)":[5000],
           "Cells":[3],
           "Discharge Rate":[20],
           "Volume (mm^3)":[188500],
           "Weight (g)":[404]})
df = df.append(aux)
```

```python
df["Volumetric Density"] = df["Capacity (mAh)"]/df["Volume (mm^3)"]
df["Weight-Charge Density"] = df["Capacity (mAh)"]/df["Weight (g)"]
df
```

```python
df.sort_values(["Volumetric Density"], ascending=False)
```
